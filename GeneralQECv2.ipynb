```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import circuit_drawer
import numpy as np
from qiskit_aer.noise import NoiseModel, pauli_error

# N = 2n + 1
N = 3

# Single qubit gate error rate
p1 = 0.001
# theta  = 2 * np.arcsin(np.sqrt(p))
# Define the bit-flip channel
bitflip1 = pauli_error([('X', p1), ('I', 1 - p1)])

# Two qubit gate error rate
p2 = 0.03
bitflip2 = pauli_error([
    ('II', (1 - p2)**2),
    ('XI', p2 * (1 - p2)),
    ('IX', p2 * (1 - p2)),
    ('XX', p2**2),
])

# Important functions
def gray_to_binary(g: int) -> int:
    x = g
    while g:
        g >>= 1
        x ^= g
    return x

def majority_bit(x: int, n: int) -> int:
    x_n = x & ((1 << n) - 1)
    return x_n.bit_count() > n // 2

def majority_str(bitstring: str) -> int:
    return bitstring.count('1') > len(bitstring) // 2

# 1. Setup Registers
# We use N qubits for the code, N - 1 ancilla qubits, N - 1 ancilla classical bits, and N - 1 + 1 classical bits for the result
q = QuantumRegister(2*N - 1, 'qubit')
c = ClassicalRegister(N, 'bit')
qc = QuantumCircuit(q, c)

# You can also play with the following
# qc.x(q[0]) sets initial state to |1>
# qc.h(q[0]) sets initial state to (|0> + |1>) / sqrt(2)

# --- STEP 1: ENCODING ---
# Spread the state to qubit 1 and 2 (Encoding |1> into |111>)
for i in range(1, N):
    qc.cx(q[0], q[i])
qc.barrier()

# --- STEP 2: INTRODUCE ERROR ---
# We simulate an error by flipping qubit 1. 
# You can change this to q[0] or q[2] and the code will still work!
# for i in range(N//2):
#     qc.x(q[i])
# qc.barrier()
# for i in range(N):
#     qc.ry(theta, q[i])
# qc.barrier

# --- STEP 3: DETECTION & CORRECTION ---
# Heart of the correction code
for i in range(N - 1):
    qc.cx(q[i], q[i + N])
    qc.cx(q[i + 1], q[i + N])
qc.barrier()

for i in range(N - 1):
    qc.measure(q[i + N], c[i])
qc.barrier()

for i in range(1, 2 ** (N - 1)):
    with qc.if_test((c, i)):
        num = gray_to_binary(i)
        if(majority_bit(num, N) == 1):
            num = ~num
        for j in range(N):
            if(num % 2 == 1):
                qc.x(q[j])
            num >>= 1
qc.barrier()

# --- STEP 4: MEASUREMENT ---
# We measure the corrected logical qubits
for i in range(N):
    qc.measure(q[i], c[i])

noise_model = NoiseModel()

# Apply the error after every gate on every qubit
noise_model.add_all_qubit_quantum_error(bitflip1, ['id', 'u', 'x', 'h'])
noise_model.add_all_qubit_quantum_error(bitflip2, ['cx'])

# --- EXECUTION ---
simulator = AerSimulator(noise_model=noise_model)
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()
counts = result.get_counts()

logical_counts = {0: 0, 1: 0}

for bits, freq in counts.items():
    logical_bit = majority_str(bits)
    logical_counts[int(logical_bit)] += freq

print("Circuit Summary:")
# print(qc.draw(output='text'))
print(f"\nResults (Target state was 0): {counts}")
print(f"\nLogical bit results: {logical_counts}")
qc.draw("mpl")
```
